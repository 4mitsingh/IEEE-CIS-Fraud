## Self-defined simple tools for Validation
### 5-fold validation

```python
def lgb_model_validation(df,  features, test_df = None ):
    
    params = {'num_leaves': 256,
          'min_child_samples': 50,
          'objective': 'binary',
#           'max_depth': 13,
          'learning_rate': 0.005,
          "boosting_type": "gbdt",
          "subsample_freq": 5,
          "subsample": 0.9,
          "bagging_seed": 11,
          "metric": 'auc',
          "verbosity": -1,
          'reg_alpha': 0.3,
          'reg_lambda': 0.3,
          'colsample_bytree': 0.9, 
          'seed':100,
         } 
#     params = {'num_leaves': 256,
#           'min_child_samples': 79,
#           'objective': 'binary',
#           'max_depth': 13,
#           'learning_rate': 0.03,
#           "boosting_type": "gbdt",
#           "subsample_freq": 3,
#           "subsample": 0.9,
#           "bagging_seed": 11,
#           "metric": 'auc',
#           "verbosity": -1,
#           'reg_alpha': 0.3,
#           'reg_lambda': 0.3,
#           'colsample_bytree': 0.9, 
#          } 

    MAX_ROUNDS = 10000
    val_pred = []
    test_pred = []
    cate_vars = []
    feature_importance = None
    models = [] 
    
    kf = StratifiedKFold(n_splits=5, shuffle=False, random_state=123)
    
    X_train = df[features].copy()
    y_train = df['isFraud'].values
    
    for fold, (tr_idx, val_idx) in enumerate(kf.split(X_train, y_train)):
        train_df,train_label = X_train.iloc[tr_idx, :].copy() ,y_train[tr_idx]
        val_df,val_label   = X_train.iloc[val_idx, :].copy() ,y_train[val_idx]
        
#         train_label  = train_df['isFraud'].values
#         val_label    = val_df['isFraud'].values# df_train_label  

        dtrain = lgb.Dataset(train_df, label = train_label)
        dval   = lgb.Dataset(val_df,   label=val_label, reference=dtrain)


        bst = lgb.train(params, dtrain, num_boost_round=MAX_ROUNDS,valid_sets=[dtrain, dval],  early_stopping_rounds=150, verbose_eval=50)
        models.append(bst)
        f_importance = pd.DataFrame()
        f_importance['fea'] = features
        f_importance['imp'] = bst.feature_importance("gain")
        f_importance['fold'] = fold
        if feature_importance is None:
            feature_importance = f_importance
        else:
            feature_importance = pd.concat([feature_importance, f_importance],axis=0,ignore_index=True)

        val_pred.append(bst.predict(
            val_df[features], num_iteration=bst.best_iteration or MAX_ROUNDS))

        if test_df is not None:
            test_pred.append(bst.predict(
                test_df[features], num_iteration=bst.best_iteration or MAX_ROUNDS))
    if test_df is None:
        return models, feature_importance, val_pred
    else:
        return models, feature_importance, val_pred, test_pred
```

```python
def fold5_lgb_model(X_train,  y_train, weights, test_df = None ):
    
    params = {'num_leaves': 256,
          'min_child_samples': 50,
          'objective': 'binary',
#           'max_depth': 13,
          'learning_rate': 0.01,
          "boosting_type": "gbdt",
          "subsample_freq": 5,
          "subsample": 0.9,
          "bagging_seed": 11,
          "metric": 'auc',
          "verbosity": -1,
          'reg_alpha': 0.3,
          'reg_lambda': 0.3,
          'colsample_bytree': 0.9, 
          'seed':100,
         } 
    MAX_ROUNDS = 2000
    test_pred = []
    
    kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=123)
     
    models = []
    for fold, (tr_idx, val_idx) in enumerate(kf.split(X_train, y_train)):
        train_df,train_label,train_weight = X_train.iloc[tr_idx, :].copy() ,y_train[tr_idx], weights[tr_idx]
        
        val_df,val_label   = X_train.iloc[val_idx, :].copy() ,y_train[val_idx] 

        dtrain = lgb.Dataset(train_df, label = train_label, weight=train_weight)
        dval   = lgb.Dataset(val_df,   label=val_label, reference=dtrain)
        bst = lgb.train(params, dtrain, num_boost_round=MAX_ROUNDS,valid_sets=[dtrain, dval],  early_stopping_rounds=150, verbose_eval=50)
        
        models.append(bst)
        if test_df is not None:
            test_pred.append(bst.predict(
                test_df, num_iteration=bst.best_iteration or MAX_ROUNDS))
    return  models,np.mean(test_pred,axis=0)

def lgb_model_test_weight_timebased(df,  features, ws = [1.01], test_df = None ): 
 
    test_pred = []
    cate_vars = []
    feature_importance = None
    models = []  
    
    tr_idx  = df['isFraud'] >=0
    te_idx  = df['isFraud'] < 0
    for i,w in enumerate(ws):
    
        train_weight = []
        tmp = df.loc[tr_idx, 'TransactionDT_day'].values
        for d in tmp:
            train_weight.append(w ** (d // 10)) 
            
        train_df,train_label = df.loc[tr_idx, features].copy() ,df.loc[tr_idx,'isFraud'].values
        
        te_df                = df.loc[te_idx, features].copy() 

        model,pred = fold5_lgb_model(train_df,train_label, np.array(train_weight), te_df) 
        models.append(model)
        test_pred.append(pred)
    return models, test_pred
```
